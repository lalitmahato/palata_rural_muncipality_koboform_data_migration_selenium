```python
# click on show more btn
def join_with(data):
    final_join = []
    for c in data:
        name = c.split()
        final_join.append("_".join(name))
    return final_join
        
def get_url(url):
    driver.get(url)
    time.sleep(8)

def show_more_btn():
    show_more = driver.find_element(By.CLASS_NAME, 'Buttons__Button-sc-19xdot-1')
    show_more.click()

def to_bs4_object(html_page):
    soup = BeautifulSoup(html_page, "html.parser")
    return soup

# Return the current url html code
def get_page_source():
    result = driver.page_source
    soup = to_bs4_object(result)
    return soup

def submit_form():
    submit = driver.find_element(By.ID, 'submit-form')
    submit.click()
    time.sleep(15)

def fill_form(form_name, data, prefix=''):
    # fill the form
    try:
        text_input = driver.find_element(By.NAME, prefix + form_name)
        text_input.clear()
        text_input.send_keys(data)
        text_input.send_keys(Keys.ENTER)
    except:
        ...

def click_radio_btn(form_name):
    radio_btn = driver.find_element(By.XPATH, form_name)
    radio_btn.click()

def select_option_value(form_name, data):
    select = Select(driver.find_element(By.NAME, form_name))
    select.select_by_value(data)

def select_dropdown(form_name, data):
    # select the dropdown
    select = Select(driver.find_element(By.NAME, form_name))
    select.select_by_visible_text(data)
    

def click_btn(value):
    btn = driver.find_element(By.CLASS_NAME, value)
    btn.click()
    
def click_btns_xpath(xpath, radio_xpath):
    try:
        time.sleep(1)
        btn = driver.find_element(By.XPATH, xpath)
        btn.click()
        time.sleep(1)
        radio_btn = driver.find_element(By.XPATH, radio_xpath)
        radio_btn.click()
    except:
        time.sleep(15)
        btn = driver.find_element(By.XPATH, xpath)
        btn.click()
        time.sleep(1)
        radio_btn = driver.find_element(By.XPATH, radio_xpath)
        radio_btn.click()
    
def click_btn_by_xpath(xpath):
    btn = driver.find_element(By.XPATH, xpath)
    btn.click()
```

```python
def click_radio_button_value(name, value, index='1'):
    try:
        xpath = f"(//input[@name='{name}'][@value='{value}'])[{index}]"
        radio_btn = driver.find_element(By.XPATH, xpath)
        radio_btn.click()
    except:
        pass
```

```python
def tick_multiple_select_value(name, value, index='1'):
    try:
        xpath = f"(//input[@name='{name}'][@value='{value}'])[{index}]"
        radio_btn = driver.find_element(By.XPATH, xpath)
        radio_btn.click()
    except:
        pass
```

```python
def fill_group_form_xpath(xpath, value):
    try:
        text_input = driver.find_element(By.XPATH, xpath)
        text_input.clear()
        text_input.send_keys(value)
    except:
        pass
```

```python
def press_group_btn(name):
    try:
        xpath = f"//div[@data-name='{name}']/button"
        rep_grp_btn = driver.find_element(By.XPATH, xpath)
        rep_grp_btn.click()
    except:
        pass
```

```python
def click_group_radio_button_value(name, value, index='1'):
    try:
        xpath = f"(//input[@data-name='{name}'][@value='{value}'])[{index}]"
        radio_btn = driver.find_element(By.XPATH, xpath)
        radio_btn.click()
    except:
        pass
```

```python
def get_image_url(attachment, value):
    for at in attachment:
        full_url = at['filename']
        file_name = full_url.split('/')[-1]
        if file_name == value:
            curret_dir = os.getcwd()
            full_url = curret_dir +'/'+ full_url
            return full_url
    return ''
```

```python
def upload_image(xpath, url):
    file_input = driver.find_element(By.XPATH, xpath)
    file_input.send_keys(url)
```
